# Webscrapping IMDb

```python
list_movie = []
list_url = []
list_year = []
list_duration = []
list_genres = []
list_storyline = []
list_keywords = []

list_country = []
list_language = []

list_directors = []
list_directors_id = []
list_directors_like = []

list_actor_1 = []
list_actor_1_id = []
list_actor_1_like = []
list_actor_2 = []
list_actor_2_id = []
list_actor_2_like = []
list_actor_3 = []
list_actor_3_id = []
list_actor_3_like = []

list_budget = []
list_gross = []
list_opening_weekend = []

list_imdb_rating = []
list_imdb_critics = []
```

```python
n = len(movies_url)
```

```python
for i in range(0,n):
    print(i)
    url = movies_url[i]['movie_imdb_link']
    source = requests.get(url).content
    soup = bs.BeautifulSoup(source,features='lxml')
    response = HtmlResponse(url=url, body = source)
    
    #basic stuff about the movie
    ## the name of the movie
    list_movie.append(movies_url[i]['movie_name'])
    ## the url of the movie
    list_url.append(url)
    ## the year of the movie
    try:
        list_year.append(response.xpath('//*[@id="titleYear"]/a/text()').extract()[0])
    except:
        list_year.append(None)
    ## the duration of the movie
    try:
        list_duration.append(response.xpath('//time[@itemprop="duration"]/text()').extract())
    except:
        list_duration.append(None)
    ## the genres of the movie
    try:
        list_genres.append(response.xpath('//div[@itemprop="genre"]//a/text()').extract())
    except:
        list_genres.append(None)
    ## the storyline of the movie
    try:
        list_storyline.append(response.xpath('//div[@id="titleStoryLine"]/div[@itemprop="description"]/p/text()').extract()[0])
    except:
        list_storyline.append(None)
    ## the keywords of the movie
    try:
        list_keywords.append(response.xpath('//a/span[@itemprop="keywords"]/text()').extract())
    except:
        list_keywords.append(None)
    
    # extra informations on the movie 
    ## the coutry of the movie
    try:
        list_country.append(response.xpath('//div[@id="titleDetails"]/div/a[contains(@href, "country")]/text()').extract()[0])
    except:
        list_country.append(None)
    ## the language of the movie
    try:
        list_language.append(response.xpath('//div[@id="titleDetails"]/div/a[contains(@href, "language")]/text()').extract()[0])
    except:
        list_language.append(None)
    ### the movie_id
    #movie_id = soup.find(class_= "imdbRatingPlugin imdbRatingStyle1").get('data-title')
    #list_movie_id.append(movie_id)
    ## the number of facebook likes of the movie
    #url_movie = "https://www.facebook.com/widgets/like.php?width=280&show_faces=1&layout=standard&href=http%3A%2F%2Fwww.imdb.com%2Ftitle%2F{}%2F&colorscheme=light".format(movie_id)
    #content = urllib.request.urlopen(url_movie).read()
    #soup_movie = BeautifulSoup(content, "lxml")
    #sentence = soup_movie.find_all(id="u_0_2")[0].text
    #list_like_movie.append(sentence.split(' ')[2][:-1])
    
    # informations about the director
    ## the name of the director
    try:
        list_directors.append(soup.find( itemprop='director').text)
    except:
        list_directors.append(None)
    ## the url to director page
    try:
        director_id = soup.find( itemprop='director').a.get('href')[6:15]
        list_directors_id.append(director_id)
    except:
        list_directors_id.append(None)
    ## the number of likes for the director
    try:
        url_director = "https://www.facebook.com/widgets/like.php?width=280&show_faces=1&layout=standard&href=http%3A%2F%2Fwww.imdb.com%2Fname%2F{}%2F&colorscheme=light".format(director_id)
        content_director = urllib.request.urlopen(url_director).read()
        soup_director = BeautifulSoup(content_director, "lxml")
        sentence_director = soup_director.find_all(id="u_0_2")[0].text
        list_directors_like.append(sentence_director)
    except:
        list_directors_like.append(None)
    
    # informations about the actors
    list_temp_name = []
    list_temp_id = []
    for temp in soup.find_all(itemprop = "actors"):
        list_temp_name.append(temp.find(itemprop = "name").text)
        list_temp_id.append(temp.a.get('href')[6:15])
    
    if len(list_temp_name) >= 3:
        list_actor_1.append(list_temp_name[0])
        list_actor_1_id.append(list_temp_id[0])
        list_actor_2.append(list_temp_name[1])
        list_actor_2_id.append(list_temp_id[1])
        list_actor_3.append(list_temp_name[2])
        list_actor_3_id.append(list_temp_id[2])
    
    if len(list_temp_name) == 2:
        list_actor_1.append(list_temp_name[0])
        list_actor_1_id.append(list_temp_id[0])
        list_actor_2.append(list_temp_name[1])
        list_actor_2_id.append(list_temp_id[1])
        list_actor_3.append(None)
        list_actor_3_id.append(None)
    
    if len(list_temp_name) == 1:
        list_actor_1.append(list_temp_name[0])
        list_actor_1_id.append(list_temp_id[0])
        list_actor_2.append(None)
        list_actor_2_id.append(None)
        list_actor_3.append(None)
        list_actor_3_id.append(None)
        
    if len(list_temp_name) == 0:
        list_actor_1.append(None)
        list_actor_1_id.append(None)
        list_actor_2.append(None)
        list_actor_2_id.append(None)
        list_actor_3.append(None)
        list_actor_3_id.append(None)
    
    try:
        url_actor = "https://www.facebook.com/widgets/like.php?width=280&show_faces=1&layout=standard&href=http%3A%2F%2Fwww.imdb.com%2Fname%2F{}%2F&colorscheme=light".format(list_temp_id[0])
        content_actor = urllib.request.urlopen(url_actor).read()
        soup_actor = BeautifulSoup(content_actor, "lxml")
        sentence_actor = soup_actor.find_all(id="u_0_2")[0].text
        list_actor_1_like.append(sentence_actor)
    except:
        list_actor_1_like.append(None)
    
    try:
        url_actor_2 = "https://www.facebook.com/widgets/like.php?width=280&show_faces=1&layout=standard&href=http%3A%2F%2Fwww.imdb.com%2Fname%2F{}%2F&colorscheme=light".format(list_temp_id[1])
        content_actor_2 = urllib.request.urlopen(url_actor_2).read()
        soup_actor_2 = BeautifulSoup(content_actor_2, "lxml")
        sentence_actor_2 = soup_actor_2.find_all(id="u_0_2")[0].text
        list_actor_2_like.append(sentence_actor_2)
    except:
        list_actor_2_like.append(None)
    
    try:
        url_actor_3 = "https://www.facebook.com/widgets/like.php?width=280&show_faces=1&layout=standard&href=http%3A%2F%2Fwww.imdb.com%2Fname%2F{}%2F&colorscheme=light".format(list_temp_id[2])
        content_actor_3 = urllib.request.urlopen(url_actor_3).read()
        soup_actor_3 = BeautifulSoup(content_actor_3, "lxml")
        sentence_actor_3 = soup_actor_3.find_all(id="u_0_2")[0].text
        list_actor_3_like.append(sentence_actor_3)
    except:
        list_actor_3_like.append(None)
    
    # informations about the money
    ## budget of the movie
    try:
        budget = response.xpath('//h4[contains(text(), "Budget:")]/following-sibling::node()/descendant-or-self::text()').extract()
        list_budget.append(budget)
    except:
        list_budget.append(None)
    ## gross of the movie
    try:
        gross = response.xpath('//h4[contains(text(), "Gross:")]/following-sibling::node()/descendant-or-self::text()').extract()
        list_gross.append(gross)
    except:
        list_gross.append(None)
    ## opening week end results
    try:
        opening = response.xpath('//h4[contains(text(), "Opening Weekend USA:")]/following-sibling::node()/descendant-or-self::text()').extract()
        list_opening_weekend.append(opening)
    except:
        list_opening_weekend.append(None)
     
    # informations about imdb rating
    try:
        list_imdb_rating.append(response.xpath("//span[@itemprop='ratingValue']/text()").extract())
    except:
        list_imdb_rating.append(None)
    try:
        list_imdb_critics.append(response.xpath('//span[@itemprop="ratingCount"]/text()').extract()[0])
    except:
        list_imdb_critics.append(None)
```

```python
def before(value, a):
    # Find first part and return slice before it.
    pos_a = value.find(a)
    if pos_a == -1: return ""
    return value[0:pos_a-1]
```

```python
def after(value, a):
    # Find and validate first part.
    pos_a = value.rfind(a)
    if pos_a == -1: return ""
    # Returns chars after the found string.
    adjusted_pos_a = pos_a + len(a)
    if adjusted_pos_a >= len(value): return ""
    return value[adjusted_pos_a+1:]
```

```python
def find_likes(x):
    try :
        likes = after(before(x, 'other'), 'and')
    except :
        likes = None
    
    if likes =='':
        return None
    else:
        return likes
```

```python
final_data.loc[:, 'Actor_1_like_temp'] = final_data['Actor_1_like'].map(lambda x: find_likes(x))
```

```python
def value_to_float(x):
    if x == None:
        return None
    #if type(x) == float or type(x) == int:
     #   return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    try:
        y = re.findall(r'-?\d+\.?\d*', x)[0]
        return int(y)
    except:
        return 0
```

```python
final_data.loc[:, 'Actor_1_like_final'] = final_data['Actor_1_like_temp'].map(lambda x: value_to_float(x))
final_data['Actor_1_like_final'] = final_data['Actor_1_like_final'].astype(float)
```
